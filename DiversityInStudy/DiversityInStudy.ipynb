## Unit Testing

In general, you should create your own tests for your code before submitting it.

[TechTarget](https://www.techtarget.com/searchsoftwarequality/definition/unit-testing) provides a good definition of unit testing from our perspective:

> Unit testing is a software development process in which the smallest testable parts of an application, called units, are individually and independently scrutinized for proper operation. This testing methodology is done during the development process by the software developers and sometimes QA staff.  The main objective of unit testing is to isolate written code to test and determine if it works as intended.
>
> Unit testing is an important step in the development process, because if done correctly, it can help detect early flaws in code which may be more difficult to find in later testing stages.

In this course we will use a compositional approach (or, looking from the other direction, _problem decomponsition_) for problem solving (and programming). This means we will use functions (or "functional units") which each have a specific, well-defined, purpose, and compose those to form a problem solution. As the above quote suggests, it is much easier to find bugs by testing a stand-alone function than it is once it is composed into a bigger program.

**Examples**

So what might a unit test look like? Normally it would be something where you can verify that the code does what you expect. In this case, one example would be to simply pick a course where you can easily count the occurrences yourself by searching through the input file. Something like CITS1021 occurs many times, so its easy to get the (manual) count wrong. So try something that is a bit rarer.

* Check your code on MUSC2275. Use 'find' in the text file to manually check that the number returned is correct.

**Edge Cases (or Limit Cases)**

Another common way of checking code is to check for "_edge cases_" - cases at the extremes. For this example, one edge case would be a unit code that doesn't appear at all. You should check that this returns zero. Another would be to write code to find the unit code that appears in the largest number of student study plans, and check that this is not greater than the total number of students (that is, lines in the input). If it is, you know there must be something wrong with the code.

* Write unit tests for each of the above test cases.

**Reductio ad Absurdum?**

When testing (and debugging) code, a useful technique is to ask what is the smallest example you can test it (or break it - see below) using. Let's say we want to sort the list of study patterns. Its very difficult for a human to check the answer when there may be hundreds of units or unit sets. However entirely equivalent (from a sorting perspective, not a cleaning perspective) cases can be designed just using letters such as `a`, `b`, `c`,... 

**White Hat mentality**

Another way of thinking about unit tests is to think about intentionally trying to "break" your code.  What cases might cause it difficulty? In the sorting example, can you come up with combinations of the letters that will in which your sort will fail?

In fact, on a larger scale, this is what a "White Hat Hacker" does, and can make you [a lot of money](https://totalcoin.io/articles/183-24-500-how-do-white-hat-hackers-cash-in-on-crypto.html)!
